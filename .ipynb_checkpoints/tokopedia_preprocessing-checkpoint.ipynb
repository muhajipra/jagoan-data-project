```python
import pandas as pd
import json
```

```python
def read_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        return data
```

```python
data = read_json('tokopedia_scraped_data.json')
```

```python
dataset = data['product info']
```

```python
for i, data in enumerate(dataset):
    try:
        data['info'] = data['info'].split('â€¢')
        for info in data['info']:
            if 'Terjual' in info:
                info = info.replace('Terjual ', '')
                info = info.replace('+', '')
                if 'rb' in info:
                    info = info.replace(' rb', '000')
                data['terjual'] = int(info)
            elif 'rating' in info:
                data['rating'], data['jumlah_ulasan'] = info.split(' (')
                data['rating'] = float(data['rating'].replace('rating ', ''))
                data['jumlah_ulasan'] = data['jumlah_ulasan'].replace(' ulasan)', '')
            elif 'Diskusi' in info:
                info = info.replace('Diskusi (', '')
                info = info.replace(')', '')
                data['diskusi'] = info
        data['harga'] = data['harga'].replace('Rp', '')
    except:
        print(i, data)
    
```

```python
df = pd.DataFrame(dataset)
df = df.drop(columns=['info'])
```

```python
print(df['asal_toko'].unique())
```

```python
df['jenis_daerah'] = 'Kabupaten' 
```

```python
df_copy = df.copy()
```

```python
df_copy['jenis_daerah'].loc[df_copy['asal_toko'].str.startswith('Jakarta')] = 'Kota' 
```

```python
df_copy['jenis_daerah'].loc[df_copy['asal_toko'].str.startswith('Kota')] = 'Kota'
df_copy['jenis_daerah'].loc[df_copy['asal_toko'].str.startswith('Kab.')] = 'Kabupaten'
```

```python
for i in df_copy.index:
    asal_daerah = df_copy.iloc[i, 5]
    if asal_daerah.startswith('Jakarta'):
        df_copy.iloc[i, -1] = 'Kota'
    elif asal_daerah.startswith('Kota'):
        df_copy.iloc[i, 5] = asal_daerah.replace('Kota ', '')
        df_copy.iloc[i, -1] = 'Kota'
    elif asal_daerah.startswith('Kab.'):
        df_copy.iloc[i, 5] = asal_daerah.replace('Kab. ', '')
        df_copy.iloc[i, -1] = 'Kabupaten'
```

```python
cols = list(df_copy.columns.values)
print(cols)
```

```python
df_copy = df_copy[['judul', 'kategori', 'harga', 'terjual', 'rating', 'jumlah_ulasan', 'diskusi', 'nama_toko', 'status_toko', 'asal_toko', 'jenis_daerah']]
```
